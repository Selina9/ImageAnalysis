# Comet analysis based on DBSCAN

Look at images of cells that have been irradiated using protons and X-rays, as well as unirradiated (control) cells. 

To produce the images, HeLa (and other) cells are embedded in 1% agarose and undergo lysis â€“ the cell membrane is broken down. Under electrophoresis, the DNA is stretched out along the electric field. More elongated "comet-like" shapes indicate more damage to the DNA. Different chemical treatments of the cells (acidic, neutral and alkaline environments) emphasize effects of different types of DNA damage. See [Jason's slides](../Talks/RadiobiologyAndDNAdamageResponse.pptx) or [Selina's talk to the Radiobiology group](../RadiobiologyGroup/RadioBiologyMeeting171201.pdf) for more information.
